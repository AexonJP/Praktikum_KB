```python
import matplotlib.pyplot as plt
# df
# datax =
datax = df.loc[0:,["Vehicle Class","Engine Size (L)","Cylinders","Fuel Consumption (City) (L/100 km)","Fuel Consumption (Hwy) (L/100 km)","Fuel Consumption (Comb) (L/100 km)","Fuel Consumption (Comb) (mpg)","CO2 Emissions (g/km)"]]
datap= datax.groupby("Vehicle Class").mean()
```

membuang data yang bukan nomor dan hanya memakai data yang tidak null

```python
# you might have to increase the size of the figure
plt.figure(figsize=(45, 10))  # width:10, height:8


x= datap.index
y = datap.Cylinders

plt.bar(x, y, align='center', width=0.2)



plt.title("rata-rata cilinder kendaraan berdasarkan kelas")
plt.xlabel('kelas kendaraan')
plt.ylabel("rata-rata cilinder kendaraan")
plt.show()
```

ini untuk menghitung rata-rata cilinder yang di miliki kendaranan berdasarkan kelasnya dalam bentuk bar

```python
x = datap.hist(bins=15, figsize=(20, 15), color='skyblue')
plt.show()
```

disini menampilkan histogram dari baris yang ada di data

```python
datam = df.loc[0:,"Vehicle Class"].value_counts()
# datam
plt.pie(datam, labels=datap.index, autopct="%1.1f%%")
plt.title('Jumlah Data')
plt.show()
```

disini menampilkan banyak data yang dimiliki oleh setiap kelas kendaraan

```python
import seaborn as sns
cor_matrix = datap.corr()
sns.heatmap(cor_matrix, cmap='coolwarm', annot=True, center=0)
plt.show()
```

disini menampilkan heatmap dari setiap kolom yang ada di data

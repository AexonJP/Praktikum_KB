```python
from tabulate import tabulate
dictBaru = {
    "banyak record" : [len(df)],
    "banyak atribute" : [len(df.columns)],
    "nama atribute" : ["\n".join(list(df.columns))],
    "banyak atribut angka" : [len(df.dtypes[df.dtypes == "float64"])],
}

print(tabulate(pd.DataFrame(dictBaru), headers="keys", tablefmt="psql", showindex=False))
```
